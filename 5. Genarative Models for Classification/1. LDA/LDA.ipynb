# Import Libraries 

```python
import numpy as np
import pandas as pd
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
import matplotlib.pyplot as plt
```

# Load Datasets

```python
animal_df1 = pd.read_csv("./animals_1.csv")
animal_df2 = pd.read_csv("./animals_2.csv")
```

# LDA where p = 1

## 1. FIT MODEL

```python
# Perform Linear Discriminant Analysis (LDA) with one components
lda = LinearDiscriminantAnalysis(n_components=1)
X = animal_df1[['Weight', 'Size']]
y = animal_df1['Class']
lda_result = lda.fit_transform(X, y)


# Add the LDA results as a new column in the dataframe
animal_df1['LDA1'] = lda_result
```

## 2. PLOT RESULT

```python

# Plot the LDA results in 1D
plt.figure(figsize=(8, 6))
colors = {'tiger': 'orange','panda': 'black'}
for animal_class, color in colors.items():
    plt.scatter(animal_df1[animal_df1['Class'] == animal_class]['LDA1'],
                np.zeros_like(animal_df1[animal_df1['Class'] == animal_class]['LDA1']),
                label=animal_class, color=color, alpha=0.7)

plt.title('LDA Projection of Panda and Tiger Data (1D)')
plt.legend()
plt.xlabel('LDA Component 1')
plt.show()
```

# LDA where p > 2
i.e we have more than 2 classes

## 1. Fit Model

```python

# Perform Linear Discriminant Analysis (LDA) with two components
lda = LinearDiscriminantAnalysis(n_components=2)
X = animal_df2[['Weight', 'Size']]
y = animal_df2['Class']
lda_result = lda.fit_transform(X, y)

# Add the LDA results as new columns in the dataframe
animal_df2['LDA1'] = lda_result[:, 0]
animal_df2['LDA2'] = lda_result[:, 1]
```

## Plot Result

```python
# Plot the LDA results in 2D
plt.figure(figsize=(8, 6))
colors = {'panda': 'black', 'tiger': 'darkorange', 'cat': 'green'}
for animal_class, color in colors.items():
    plt.scatter(animal_df2[animal_df2['Class'] == animal_class]['LDA1'],
                animal_df2[animal_df2['Class'] == animal_class]['LDA2'],
                label=animal_class, color=color, alpha=0.7)

plt.title('LDA Projection of Animal Data (2D)')
plt.legend()
plt.xlabel('LDA Component 1')
plt.ylabel('LDA Component 2')
plt.show()
```
